**State wise sales, discount and profit analysis**

```
df1=df.groupby(['State'])[['Sales','Profit','Discount']].mean()
df1.head()
```

```
df2 =  df['State'].value_counts().to_frame()
df1 = pd.merge(df1, df2, left_index=True, right_index=True)
df1.rename(columns={'State':'No_of_deals'},inplace=True)
df1.head()
```

```
plt.figure(figsize = (25,5))
sns.violinplot(data = df, x='State', y= 'Sales',palette='muted')
plt.xticks(rotation = 90)
plt.ylabel('Average Sales in Dollars')
plt.xlabel('States')
plt.show()
```

```
df1.sort_values('Profit')[['No_of_deals','Sales','Profit']].plot(kind='bar', figsize= (25,5),label=['No_of_deals','Avg Sales Price($)','Profit($)'])
plt.ylabel('No Of Deals/Average Sales Price($) and Profit per State($)')
plt.xlabel('States')
plt.legend()
plt.show()
```

```
df1.sort_values('Profit')[['Discount','Profit']].plot(kind='bar', figsize= (25,5))
plt.ylabel('Discount($) and Profit($) per State')
plt.xlabel('States')
plt.legend()
plt.show()
```

We can see we can get maximum profit in Vermont State and Minimum in Ohio.

In Vermont No of Deals is very less still Sales in Dollars and Profit is much more. This indicates the quality of the deals is too good. (Maybe the expensive goods are bought more at Vermont)

In Ohio Profit is negative and loss is maximum. Issues faced in Ohio must be addressed and solve to perform better.
